### Unsupervised Learning Exercise: Vehicle Clustering

**Scenario** - You are working for an automotive company, and your task is to cluster vehicles into groups based on their features such as weight, engine size, and horsepower.

In this exercise, you used KMeans clustering to group vehicles based on their features like weight, engine size, and horsepower.

```python
# Import necessary libraries
import pandas as pd  # For handling data in DataFrames
import numpy as np  # For generating random data
import matplotlib.pyplot as plt  # For plotting data
import warnings  # To handle warnings
from sklearn.cluster import KMeans  # KMeans algorithm for clustering
```

```python
# Ignore warnings to keep the output clean
warnings.filterwarnings('ignore')

# Generate synthetic dataset for vehicles
np.random.seed(0)  # Set the seed for reproducibility
data_size = 300  # Define the number of data points (vehicles)
```

```python
# Create a dictionary with random values for vehicle attributes
data = {
    'Weight': np.random.randint(1000, 3000, data_size),  # Random integer values for vehicle weight (range: 1000-3000 kg)
    'EngineSize': np.random.uniform(1.0, 4.0, data_size),  # Random float values for engine size (range: 1.0-4.0 liters)
    'Horsepower': np.random.randint(50, 300, data_size)  # Random integer values for horsepower (range: 50-300 HP)
}

print(data)
```

```python
# Create a DataFrame from the generated data
df = pd.DataFrame(data)

# No labels are needed for unsupervised learning (since it's clustering)
X = df  # Assign the DataFrame to X for clustering (features)

print(df)
```

```python
# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Initialize KMeans with 3 clusters and a random seed
kmeans.fit(X)  # Fit the KMeans model to the data
```

```python
# Plotting the clusters
# Scatter plot with 'Weight' on the x-axis and 'Horsepower' on the y-axis, colored by cluster labels
plt.scatter(df['Weight'], df['Horsepower'], c=kmeans.labels_)  # Color points based on assigned clusters
plt.xlabel('Weight')  # Label for x-axis (vehicle weight)
plt.ylabel('Horsepower')  # Label for y-axis (horsepower)
plt.title('Vehicle Clusters')  # Title of the plot
plt.show()  # Display the plot
```

#### Expected Results
1. Cluster Visualization - The scatter plot will visually depict how vehicles are grouped based on weight and horsepower. Each cluster will be represented by a different color. With synthetic data, the distinctness of clusters may vary, but students should be able to see groupings based on the chosen features.
2. Interpreting Clusters - There are no 'correct' labels in unsupervised learning, but students should observe how vehicles are grouped based on similarities in their features. They might see, for example, that heavier vehicles with higher horsepower are grouped together.
